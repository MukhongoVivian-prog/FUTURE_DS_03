## Data Preprocessing
### Handling missing values

```python
#Handling The Missing Values
data.isnull().sum()
```

```python
# Fill categorical missing values with mode
data['Gender'].fillna(data['Gender'].mode()[0], inplace=True)
data['Married'].fillna(data['Married'].mode()[0], inplace=True)
data['Dependents'].fillna(data['Dependents'].mode()[0], inplace=True)
data['Self_Employed'].fillna(data['Self_Employed'].mode()[0], inplace=True)
data['Loan_Amount_Term'].fillna(data['Loan_Amount_Term'].mode()[0], inplace=True)
data['Credit_History'].fillna(data['Credit_History'].mode()[0], inplace=True)

# Fill numerical missing values with median
data['LoanAmount'].fillna(data['LoanAmount'].median(), inplace=True)

```

```python
data
```

### Checking for missing values

```python
data.isnull().sum()
```

### Encoding Categorical Variables

```python
# Encoding Loan_Status as binary values
data['Loan_Status'] = data['Loan_Status'].apply(lambda x: 1 if x == 'Y' else 0)

# Encoding other categorical variables
categorical_cols = ['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'Property_Area']
data = pd.get_dummies(data, columns=categorical_cols, drop_first=True)

```

```python
data
```

### Feature Scaling to improve the performance of the model

```python
from sklearn.preprocessing import StandardScaler

scaler  = StandardScaler()
data[['ApplicantIncome','Coapplication','LoanAmount']]=scaler.fit_transform(data[['ApplicantIncome','CoapplicantIncome','LoanAmount']])
```

### Features and Target Variable Defining

```python
# Defining Features and Target Variable
x = data.drop(columns=['Loan_ID','Loan_Status'])
y = data['Loan_Status']
```

```python
data
```

### Splitting dataset into Training and Testing sets

```python
# splitting dataset into training and testing sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test =train_test_split(x, y, test_size =0.2, random_state =42)
```

### Using RandomForest Classifier as a starting model.

```python
# Training the model using random forest classifier as a starting model
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators = 100, random_state = 42)
model.fit(x_train, y_train)

```

### Creating the model, training the model on data and accessing feature importance

```python

from sklearn.ensemble import RandomForestClassifier

# Creating the model
model = RandomForestClassifier()
model.fit(x_train, y_train)

#  accessing feature importance
importances = model.feature_importances_
print(importance)
```

### Evaluating the model using accuracy,confusion matrix and classification report

```python

from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
y_pred = model.predict(x_test)

print("Accuracy:", accuracy_score(y_test,y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test,y_pred))
print("Classification Report:\n", classification_report(y_test,y_pred))
```

### CREATING VISUALIZATION

```python
feature_importances = pd.Series(model.feature_importances_, index=x.columns)
feature_importances.nlargest(10).plot(kind='barh')
plt.title("Top 10 Feature Importances")
plt.show()
```

## Recommendations
#### 1. Since credit History is the most important and influential factor, financial institutions should educate applicants about the importance of maintaining good credit.
#### 2. With applicants' and co-applicants' income levels being critical, lenders should consider loan products for those with stable though slightly lower incomes.
#### 3. Given the significance of the loan amount, lenders could offer smaller more manageable loans to applicants who might struggle with higher loan amounts thus reducing the risk of default and improving overall approval rates.
#### 4. Although the property area has minor importance, lenders might still evaluate regional trends and risks associated with different locations.
#### 5. Since gender and marital status have little influence on loan approval lenders could focus on objective financial metrics rather than demographic attributes.
